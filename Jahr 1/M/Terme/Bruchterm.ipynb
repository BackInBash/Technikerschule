```python
# Bruchterm
Man spricht von einem Bruchterm wenn sich eine Variable im Nenner befindet.

\begin{equation*}
\frac{1}{3+x}
\end{equation*}

# Definitionsmenge
Bruchterme 

# Kehrwert
Zähler und Nenner werden getauscht.

\begin{equation*}
\frac{5}{x-1}
\end{equation*}

\begin{equation*}
\frac{x-1}{5}
\end{equation*}

# Kürzen und Erweitern
Bruchterme lassen sich wie gewöhnliche Brüche kürzen oder erweitern.

**Beispiel Kürzen**
\begin{equation*}
\frac{64+\frac{1}{2}bc}{88+b^2c^2}
\end{equation*}

\begin{equation*}
\frac{8a}{11bc^2}
\end{equation*}

**Beispeiel Erweitern:**

\begin{equation*}
\frac{(x+y)(x+y)}{(x+y)(x-y)}
\end{equation*}

\begin{equation*}
\frac{(x+y)^2}{x^2-y^2}
\end{equation*}

# (Grund) Rechenoperationen

## Addition
+ Gleichnamige Brüche (Gleiche Nenner auf beiden Seiten)
\begin{equation*}
\frac{5}{2a} + \frac{7b}{2a} = \frac{5+7b}{2a}
\end{equation*}

+ Nicht gleichenamige Brüche (Nenner sind verschieden)

\begin{equation*}
\frac{3}{x^2} + \frac{5}{9} = \frac{5+7b}{2a}
\end{equation*}

Für Addition müssen beide Bruchterme den gleichen Nenner haben.

\begin{equation*}
\frac{a}{b}+\frac{c}{d} = \frac{a*b+c*b}{ab}
\end{equation*}

## Subtraktion

\begin{equation*}
\frac{5}{x+1}-\frac{2}{2+x} = \frac{5(x+2)-2(x+1)}{(x+1)(x+2)} = \frac{5x+10-2x-2}{x^2+2x-x+2} = \frac{3x+8}{x^2+3x+2}
\end{equation*}

## Multiplikation

+ Brüche werden multipliziert indem man die Zähler und Nenner jeweils multipliziert

\begin{equation*}
\frac{3x}{(x+y)^2} * \frac{x+y}{x-y} = \frac{3x(x+y)}{(x+y)^2(x-y)}
\end{equation*}

## Division

\begin{equation*}
\frac{\frac{a^2+1}{x}}{\frac{a+1}{x^2+b}} = \frac{x^2-1}{a} * \frac{x^2b}{a+1} = (a-1)ba
\end{equation*}

Zwei Brücher werden dividiert indem man den ersten Bruch mit dem Kerhwert des zweiten Bruchs multipliziert.
```
